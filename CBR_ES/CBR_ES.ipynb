```python
#Drop cuando el elemento tiene la estructura .Pandas
#Axis columna 1, axis 0 filas
base = library.drop([' Play'], axis = 1)
```

```python
#Se usara recuperación para muestra problema
#Get dummies representa todos los datos en combinación binaria
#0 cuando se cumpla, 
base = pd.get_dummies(base)
problems = pd.get_dummies(cases)
```

```python
#Para sacar la covarianza de las características
#variación con respecto a la media de cada característica
covariance_matrix = base.cov()
#Para el calculo de distancia se necesita la inversa de la covarianza
inverse_covariance_matrix = np.linalg.pinv(covariance_matrix)
```

```python
covariance_matrix
```

```python
inverse_covariance_matrix
```

```python
#Primer dato fila, segundo dato columna
case_row = problems.loc[0, :]
print(case_row)
```

```python
# matriz de zeros para almacenar distancias
distances = np.zeros(base.shape[0])
print(distances)
```

```python
# Calcular distncia con cada elemento de la base de datos
# toma cada elemento de la lista y lo procesa con cada uno.
for j in range(base.shape[0]):

  base_row = base.loc[j, :]

  #Calcular distancia mahalanobis
  distances[j] = distance.mahalanobis(case_row, base_row, inverse_covariance_matrix)
```

```python
distances
```

```python
# recuperar el indice con la distancia minima
# en caso de multiples toma la primera en orden ascendente de indice
min_distance_row = np.argmin(distances)
print(min_distance_row)
```

```python
#Recuperar caso sin codificar y agregar solución adquirida
# Append - te ayuda a fusionar la solución recuperada para sacar la conclusión
case = np.append(cases.iloc[0, :], library.iloc[min_distance_row, -1])
print(case)
```

```python
print(f'> Para el caso/problema [0]: {cases.iloc[0, :].to_numpy()}, la solución {case[-1]}')
```

```python
# Agregarla nueva solución a la libreria
case = pd.Series(case, index = library.columns)     #Problema con la solución

library = library.append(case, ignore_index= True)  #Actualizar libreria
```
